```python
# Load api key from .env file

from dotenv import load_dotenv

load_dotenv()

from dotenv import dotenv_values

config = dotenv_values(".env")
auth_token = config.get("ANTHROPIC_API_KEY")
```

```python
# Auth and model

from anthropic import Anthropic

client = Anthropic(
    auth_token=auth_token,
    base_url="https://flow.ciandt.com/flow-llm-proxy"
)

model = "bedrock/anthropic.claude-4-6-sonnet"
# model = "claude-sonnet-4-0"
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)

message.content[0].text
```

## 2 Multi-Turn conversations

```python
# Helper functions

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

```python
# Multi-turn conversation example

## Start with an empty message list
messages = []

## Add the initial user question
newUserMessage = "Define quantum computing in one sentence"
add_user_message(messages, newUserMessage)

## Get Claude's response and add it to the conversation history
answer = chat(messages)
add_assistant_message(messages, answer)

## Add a follow-up question
newUserMessage = "Write another sentence"
add_user_message(messages, newUserMessage)

## Get the follow-up response with full context
final_answer = chat(messages)

print(final_answer)
```
